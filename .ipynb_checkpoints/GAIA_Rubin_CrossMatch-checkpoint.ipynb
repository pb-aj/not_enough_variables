## Library and Data Setup

```python
#import libraries and set up tap service
import matplotlib.pyplot as plt
import numpy as np
from astropy.table import Table
import pyvo
import astropy.units as u
from lsst.rsp import get_tap_service
import pandas as pd
pd.set_option('display.max_columns', None)

from lsst.utils.plotting import (get_multiband_plot_colors,
                                 get_multiband_plot_symbols)

service = get_tap_service("tap")
assert service is not None
```

```python
#read in GAIA dataset filtered for probably RRL's in DP1
refined_GAIA_df = pd.read_csv("data/refined_GaiaDR3_center_FieldsDP1.csv")
refined_GAIA_df.rename(columns={'Unnamed: 0.1': 'GAIA_index'},inplace=True)
refined_GAIA_df.drop('Unnamed: 0', axis=1, inplace=True)
refined_GAIA_df.head()
```

```python
#filter pandas table and convert to astropy
#this is needed to use in TAP query!

GAIA_table_df = refined_GAIA_df[["DESIGNATION","source_id","pf","pf_error","p1_o","p1_o_error",
                                               "ra","dec","ra_error","dec_error","best_classification","field_loc_center",
                                               "phot_g_mean_flux","phot_g_mean_flux_error","phot_g_mean_mag",
                                               "phot_bp_mean_flux","phot_bp_mean_flux_error","phot_bp_mean_mag",
                                               "phot_rp_mean_flux","phot_rp_mean_flux_error","phot_rp_mean_mag",
                                               "bp_rp","bp_g","g_rp",
                                               "int_average_g","int_average_g_error",
                                               "int_average_bp","int_average_bp_error",
                                               "int_average_rp","int_average_rp_error",
                                               "peak_to_peak_g","peak_to_peak_g_error",
                                               "peak_to_peak_bp","peak_to_peak_bp_error",
                                               "peak_to_peak_rp","peak_to_peak_rp_error"]]

GAIA_table_df.columns = ["DR3_"+column for column in GAIA_table_df.columns]
GAIA_table = Table.from_pandas(GAIA_table_df[["DR3_source_id","DR3_ra","DR3_dec"]], index=False)
GAIA_table_df.head()
```

```python
#define a function to pull out forced sources for each found object
def search_for_flux(row):
    if "diaObjectId" in row.index:
        DiaObjID = row["diaObjectId"]
        query = """SELECT fsodo.coord_ra, fsodo.coord_dec,
            fsodo.diaObjectId, fsodo.visit, fsodo.band, 
            fsodo.psfDiffFlux, fsodo.psfDiffFluxErr, 
            fsodo.psfFlux as psfFlux, fsodo.psfFluxErr, 
            vis.expMidptMJD, 
            pixelFlags_bad, pixelFlags_saturated, pixelFlags_saturatedCenter,
            pixelFlags_suspect, pixelFlags_suspectCenter
            FROM dp1.ForcedSourceOnDiaObject as fsodo
            JOIN dp1.Visit as vis ON vis.visit = fsodo.visit
            WHERE fsodo.diaObjectId = """+str(DiaObjID)
    elif "ObjectId" in row.index:
        ObjID = row["ObjectId"]
        query = """SELECT fsodo.coord_ra, fsodo.coord_dec,
            fsodo.ObjectId, fsodo.visit, fsodo.band,
            fsodo.psfDiffFlux, fsodo.psfDiffFluxErr,
            fsodo.psfFlux as psfFlux, fsodo.psfFluxErr,
            vis.expMidptMJD,
            pixelFlags_bad, pixelFlags_saturated, pixelFlags_saturatedCenter,
            pixelFlags_suspect, pixelFlags_suspectCenter
            FROM dp1.ForcedSource as fsodo 
            JOIN dp1.Visit as vis ON vis.visit = fsodo.visit
            WHERE fsodo.objectId = """+str(ObjID)
    else:
        print("uh oh")
        

    job = service.submit_job(query)
    job.run()
    job.wait(phases=['COMPLETED', 'ERROR'])
    # print('Job phase is', job.phase)
    if job.phase == 'ERROR':
        job.raise_if_error()

    assert job.phase == 'COMPLETED'
    forced_source = job.fetch_result().to_table()

    forced_df = forced_source.to_pandas()

    forced_df["query"] = query
    forced_df["nSources"] = len(forced_source)

    back_sat = ~forced_df["pixelFlags_saturated"]
    cent_sat = ~forced_df["pixelFlags_saturatedCenter"]
    bad_pixel = ~forced_df["pixelFlags_bad"]

    mask = back_sat & cent_sat & bad_pixel

    forced_df["nGoodSources"] = len(forced_df[mask])
    
    for band in filter_names:
        forced_df[f"n{band}"] = np.sum((forced_df['band'] == band))
        forced_df[f"nGood{band}"] = np.sum((forced_df[mask]['band'] == band))

    return forced_df
```

```python
def move_columns(df):
    if "diaObjectId" in df.columns:
        removed_info = df[["diaObjectId","nSources","nGoodSources","query","nu",
                           "ng","nr","ni","nz","ny","nGoodu","nGoodg","nGoodr",
                           "nGoodi","nGoodz","nGoody"]].drop_duplicates()
        df.drop(columns=["nSources","nGoodSources","query","nu","ng","nr","ni","nz","ny",
                        "nGoodu","nGoodg","nGoodr","nGoodi","nGoodz","nGoody"], inplace=True)
        return removed_info
    elif "ObjectId" in df.columns:
        removed_info = df[["ObjectId","nGoodSources","nSources","query","nu",
                           "ng","nr","ni","nz","ny","nGoodu","nGoodg","nGoodr",
                           "nGoodi","nGoodz","nGoody"]].drop_duplicates()
        df.drop(columns=["nSources","query","nu","ng","nr","ni","nz","ny",
                         "nGoodu","nGoodg","nGoodr","nGoodi","nGoodz","nGoody"], inplace=True)
        return removed_info
    else:
        print("uh oh")
        return None
```

## Adding ObjId Matches to DiaObj

```python
def determine_ifin_Obj(row):

    ra,dec = row["ra"],row["dec"]

    search_range = 0.00027

    ra_mask = Obj_df["ra"].between(ra-search_range,search_range+ra)
    dec_mask = Obj_df["dec"].between(dec-search_range,search_range+dec)

    potential_id = Obj_df["ObjectId"][ra_mask & dec_mask].values

    try:
        return potential_id[0]
    except:
        return 0
```

```python
diaObj_df["ObjectId"] = diaObj_df.apply(determine_ifin_Obj,axis=1)
print(f"{86 - (diaObj_df["ObjectId"] == 0).sum()}/25 ObjIds Found in DiaObject.")
```
