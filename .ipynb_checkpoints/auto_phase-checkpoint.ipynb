# Auto Phase Curves

In this notebook, the code will go through the passed stars and save phase curve plots for each in their respective folders (diaObjectId) in auto_plots

## Importing Libraries

```python
#import libraries and set up tap service
import matplotlib.pyplot as plt
import numpy as np
from astropy.table import Table
import pyvo
import astropy.units as u
from lsst.rsp import get_tap_service
import pandas as pd
pd.set_option('display.max_columns', None)

from lsst.utils.plotting import (get_multiband_plot_colors,
                                 get_multiband_plot_symbols)

from lsst.daf.butler import Butler
from lsst.utils.plotting import (
    get_multiband_plot_colors,
    get_multiband_plot_symbols,
    get_multiband_plot_linestyles
)

service = get_tap_service("tap")
assert service is not None

import seaborn as sns
my_pal = sns.color_palette("colorblind")

import warnings
warnings.filterwarnings('error')
```

```python
Ex_diaObj_df = pd.read_csv("data/Extra_diaObj_df.csv",index_col=0)
forcedDiaObj_df = pd.read_csv("data/diaForced_df.csv",index_col=0)

grouped_forcedDia = forcedDiaObj_df.groupby("diaObjectId")

filter_names = ['u', 'g', 'r', 'i', 'z', 'y']
filter_colors = get_multiband_plot_colors()
filter_symbols = get_multiband_plot_symbols()
filter_linestyles = get_multiband_plot_linestyles()
```

## Defining Functions for Plotting

```python
def days_flux_plots(star_group, diaId, directory=""):    
    fig, ax = plt.subplots(7, 2, figsize=(14, 21))
    fig.subplots_adjust(wspace=0)
    
    ax1 = ax[0,0]
    ax2 = ax[0,1]
    for band in filter_names:
        pickband = star_group['band'] == band
        ax1.errorbar(star_group[pickband]['expMidptMJD'], star_group[pickband]['psfDiffFlux'],
                     yerr=star_group[pickband]["psfDiffFluxErr"],
                     marker=filter_symbols[band], color=filter_colors[band],
                     label=f"{band}", markersize=7, linestyle='none', fillstyle='none')
        ax2.errorbar(star_group[pickband]['expMidptMJD'], star_group[pickband]['psfFlux'],
                 yerr=star_group[pickband]["psfFluxErr"],
                 marker=filter_symbols[band], color=filter_colors[band], linestyle='none',
                 markersize=7, fillstyle='none')
    
    ax2.yaxis.tick_right()
    ax2.yaxis.set_label_position('right')

    ax1.tick_params(labelsize=10)
    ax2.tick_params(labelsize=10)
    
    ax1.set_ylabel('PSF difference\nimage flux (nJy)',fontsize=14)
    ax2.set_ylabel('PSF direct\nimage flux (nJy)',rotation=-90,labelpad=35,fontsize=14)
    ax1.set_xlabel('Exposure midpoint, MJD (d)',fontsize=14)
    ax2.set_xlabel('Exposure midpoint, MJD (d)',fontsize=14)

    ax1.set_title(f"Overall DiffFlux MJD Curve",fontsize=16)
    ax2.set_title(f"Overall Flux MJD Curve",fontsize=16)

    ax1.set_xlim(np.min(star_group['expMidptMJD']) - .5,np.max(star_group['expMidptMJD'] + .5))
    ax2.set_xlim(np.min(star_group['expMidptMJD']) - .5,np.max(star_group['expMidptMJD'] + .5))
    
    ax1.minorticks_on()
    ax2.minorticks_on()
    ax1.legend(ncols=2)
    
    for i in range(1,7):
        ax1 = ax[i,0]
        ax2 = ax[i,1]
    
        band = filter_names[i-1]
    
        pickband = star_group['band'] == band
        ax1.errorbar(star_group[pickband]['expMidptMJD'], star_group[pickband]['psfDiffFlux'],
                     yerr=star_group[pickband]["psfDiffFluxErr"],
                     marker=filter_symbols[band], color=filter_colors[band],
                     label=f"{band}", markersize=7, linestyle='none', fillstyle='none')
        ax2.errorbar(star_group[pickband]['expMidptMJD'], star_group[pickband]['psfFlux'],
                 yerr=star_group[pickband]["psfFluxErr"],
                 marker=filter_symbols[band], color=filter_colors[band], linestyle='none',
                 markersize=7, fillstyle='none')
    
        ax2.yaxis.tick_right()
        ax2.yaxis.set_label_position('right')
        
        ax1.set_ylabel('PSF difference\nimage flux (nJy)',fontsize=14)
        ax2.set_ylabel('PSF direct\nimage flux (nJy)',rotation=-90,labelpad=35,fontsize=14)
        ax1.set_xlabel('Exposure midpoint, MJD (d)',fontsize=14)
        ax2.set_xlabel('Exposure midpoint, MJD (d)',fontsize=14)

        ax1.set_title(f"{band}-band DiffFlux MJD Curve",fontsize=16)
        ax2.set_title(f"{band}-band Flux MJD Curve",fontsize=16)

        ax1.tick_params(labelsize=10)
        ax2.tick_params(labelsize=10)

        ax1.set_xlim(np.min(star_group['expMidptMJD']) - .5,np.max(star_group['expMidptMJD'] + .5))
        ax2.set_xlim(np.min(star_group['expMidptMJD']) - .5,np.max(star_group['expMidptMJD'] + .5))
        
        ax1.minorticks_on()
        ax2.minorticks_on()
        ax1.legend(ncols=2,loc="upper left")
        
    plt.tight_layout()
    plt.suptitle(f"DiaObjectId - {diaId}",fontsize=20,y=1.01) #putting after tight layout makes each more uniform
    
    if directory:
        plt.savefig(f"{directory}/combined_days.png",dpi=330,bbox_inches="tight")
        pass
    else:
        plt.show()
    plt.close()

"""
================================================================================================================
Overall Above ^

Individual Below v
================================================================================================================
"""

def individ_days_flux_plots(star_group,diaId,directory=""):
    fig, ax = plt.subplots(1, 2, figsize=(14, 4), sharex=True)
    fig.subplots_adjust(wspace=0)
    
    ax1 = ax[0]
    ax2 = ax[1]
    for band in filter_names:
        pickband = star_group['band'] == band
        ax1.errorbar(star_group[pickband]['expMidptMJD'], star_group[pickband]['psfDiffFlux'],
                     yerr=star_group[pickband]["psfDiffFluxErr"],
                     marker=filter_symbols[band], color=filter_colors[band],
                     label=f"{band}", markersize=7, linestyle='none', fillstyle='none')
        ax2.errorbar(star_group[pickband]['expMidptMJD'], star_group[pickband]['psfFlux'],
                 yerr=star_group[pickband]["psfFluxErr"],
                 marker=filter_symbols[band], color=filter_colors[band], linestyle='none',
                 markersize=7, fillstyle='none')
    
    ax2.yaxis.tick_right()
    ax2.yaxis.set_label_position('right')

    ax1.tick_params(labelsize=10)
    ax2.tick_params(labelsize=10)
    
    ax1.set_ylabel('PSF difference\nimage flux (nJy)',fontsize=14)
    ax2.set_ylabel('PSF direct\nimage flux (nJy)',rotation=-90,labelpad=35,fontsize=14)
    ax1.set_xlabel('Exposure midpoint, MJD (d)',fontsize=14)
    ax2.set_xlabel('Exposure midpoint, MJD (d)',fontsize=14)

    ax1.set_title(f"Overall DiffFlux MJD Curve",fontsize=16)
    ax2.set_title(f"Overall Flux MJD Curve",fontsize=16)

    ax1.set_xlim(np.min(star_group['expMidptMJD']) - .5,np.max(star_group['expMidptMJD'] + .5))
    ax2.set_xlim(np.min(star_group['expMidptMJD']) - .5,np.max(star_group['expMidptMJD'] + .5))
    
    ax1.minorticks_on()
    ax2.minorticks_on()
    ax1.legend(ncols=2)
    
    plt.suptitle(f"DiaObjectId - {diaId}",fontsize=20)
    plt.tight_layout()
    
    if directory:
        plt.savefig(f"{directory}/overall_days.png",dpi=330,bbox_inches="tight")
        pass
    else:
        plt.show()
    plt.close()
    
    for i in range(1,7):
        fig, ax = plt.subplots(1, 2, figsize=(14, 4), sharex=True)
        fig.subplots_adjust(wspace=0)
        
        ax1 = ax[0]
        ax2 = ax[1]
    
        band = filter_names[i-1]
    
        pickband = star_group['band'] == band
        ax1.errorbar(star_group[pickband]['expMidptMJD'], star_group[pickband]['psfDiffFlux'],
                     yerr=star_group[pickband]["psfDiffFluxErr"],
                     marker=filter_symbols[band], color=filter_colors[band],
                     label=f"{band}", markersize=7, linestyle='none', fillstyle='none')
        ax2.errorbar(star_group[pickband]['expMidptMJD'], star_group[pickband]['psfFlux'],
                 yerr=star_group[pickband]["psfFluxErr"],
                 marker=filter_symbols[band], color=filter_colors[band], linestyle='none',
                 markersize=7, fillstyle='none')
    
        ax2.yaxis.tick_right()
        ax2.yaxis.set_label_position('right')
        
        ax1.set_ylabel('PSF difference\nimage flux (nJy)',fontsize=14)
        ax2.set_ylabel('PSF direct\nimage flux (nJy)',rotation=-90,labelpad=35,fontsize=14)
        ax1.set_xlabel('Exposure midpoint, MJD (d)',fontsize=14)
        ax2.set_xlabel('Exposure midpoint, MJD (d)',fontsize=14)

        ax1.set_title(f"{band}-band DiffFlux MJD Curve",fontsize=16)
        ax2.set_title(f"{band}-band Flux MJD Curve",fontsize=16)

        ax1.tick_params(labelsize=10)
        ax2.tick_params(labelsize=10)

        ax1.set_xlim(np.min(star_group['expMidptMJD']) - .5,np.max(star_group['expMidptMJD'] + .5))
        ax2.set_xlim(np.min(star_group['expMidptMJD']) - .5,np.max(star_group['expMidptMJD'] + .5))
        
        ax1.minorticks_on()
        ax2.minorticks_on()
        ax1.legend(ncols=2,loc="upper left")
        plt.suptitle(f"DiaObjectId - {diaId}",fontsize=20)
        plt.tight_layout()
    
        if directory:
            plt.savefig(f"{directory}/{band}_days.png",dpi=330,bbox_inches="tight")            
            pass
        else:
            plt.show()
        plt.close()
```

```python
def filt_phase_flux_plots(star_group,diaId,phase,directory="",filt=2.5):
    fig, ax = plt.subplots(7, 2, figsize=(14, 21))
    fig.subplots_adjust(wspace=0)
    
    max_diff = float("-inf")
    min_diff = float("inf")
    
    max_flux = float("-inf")
    min_flux = float("inf")
    
    ax1 = ax[0,0]
    ax2 = ax[0,1]
    for band in filter_names:
        pickband = (star_group['band'] == band) #& (star_group['psfDiffFlux'] > -50000)
    
        temp_diff = star_group[pickband]['psfDiffFlux']
        temp_flux = star_group[pickband]['psfFlux']
        
        if len(temp_flux):
            med = np.median(temp_diff)
            std3 = np.std(temp_diff) * filt
        else:
            med = 0
            std3 = 0
    
        mask = (temp_diff > med - std3) & (temp_diff < med + std3)
    
        masked_diff = temp_diff[mask]
        masked_flux = temp_flux[mask]
    
        full_phase = pd.concat([phase[pickband][mask],phase[pickband][mask] + 1])
        full_diff = pd.concat([masked_diff,masked_diff])
        full_diff_err = pd.concat([star_group[pickband]['psfDiffFluxErr'][mask],star_group[pickband]['psfDiffFluxErr'][mask]])
    
        full_flux = pd.concat([masked_flux,masked_flux])
        full_flux_err = pd.concat([star_group[pickband]['psfFluxErr'][mask],star_group[pickband]['psfFluxErr'][mask]])
        
        ax1.errorbar(full_phase, full_diff,
                 yerr=full_diff_err,
                 marker=filter_symbols[band], color=filter_colors[band],
                 label=f"{band}", markersize=7, linestyle='none', fillstyle='none')
        ax2.errorbar(full_phase, full_flux,
                     yerr=full_flux_err,
                     marker=filter_symbols[band], color=filter_colors[band], linestyle='none',
                     markersize=7, fillstyle='none')
    
    
        if np.max(masked_diff) > max_diff:
            max_diff = np.max(masked_diff)
    
        if np.min(masked_diff) < min_diff:
            min_diff = np.min(masked_diff) 
    
        if np.max(masked_flux) > max_flux:
            max_flux = np.max(masked_flux)
    
        if np.min(masked_flux) < min_flux:
            min_flux = np.min(masked_flux) 
    
    ax2.yaxis.tick_right()
    ax2.yaxis.set_label_position('right')
    
    ax1.minorticks_on()
    ax2.minorticks_on()
    ax1.legend(ncols=2,loc="upper left")
    ax1.set_ylabel('PSF difference\nimage flux (nJy)',fontsize=14)
    ax2.set_ylabel('PSF direct\nimage flux (nJy)',rotation=-90,labelpad=35,fontsize=14)
    ax2.set_xlabel('Phase',fontsize=14)
    ax1.set_xlabel('Phase',fontsize=14)

    ax1.set_title(f"Overall DiffFlux Phase Curve (Filtered)",fontsize=16)
    ax2.set_title(f"Overall Flux Phase Curve (Filtered)",fontsize=16)

    try:
        ax1.set_ylim(min_diff - 2e4, max_diff + 2e4)
        ax2.set_ylim(min_flux - 2e4, max_flux + 2e4)
    except:
        pass
    
    ax1.set_xlim(-0.05,2.05)
    ax2.set_xlim(-0.05,2.05)
    
    for i in range(1,7):
        ax1 = ax[i,0]
        ax2 = ax[i,1]
    
        band = filter_names[i-1]
    
        pickband = (star_group['band'] == band) #& (star_group['psfDiffFlux'] > -50000)
    
        temp_diff = star_group[pickband]['psfDiffFlux']
        temp_flux = star_group[pickband]['psfFlux']
    
        if len(temp_flux):
            med = np.median(temp_diff)
            std3 = np.std(temp_diff) * filt
        else:
            med = 0
            std3 = 0
    
        mask = (temp_diff > med - std3) & (temp_diff < med + std3)
    
        masked_diff = temp_diff[mask]
        masked_flux = temp_flux[mask]
    
        full_phase = pd.concat([phase[pickband][mask],phase[pickband][mask] + 1])
        full_diff = pd.concat([masked_diff,masked_diff])
        full_diff_err = pd.concat([star_group[pickband]['psfDiffFluxErr'][mask],star_group[pickband]['psfDiffFluxErr'][mask]])
    
        full_flux = pd.concat([masked_flux,masked_flux])
        full_flux_err = pd.concat([star_group[pickband]['psfFluxErr'][mask],star_group[pickband]['psfFluxErr'][mask]])
        
        ax1.errorbar(full_phase, full_diff,
                 yerr=full_diff_err,
                 marker=filter_symbols[band], color=filter_colors[band],
                 label=f"{band}", markersize=7, linestyle='none', fillstyle='none')
        ax2.errorbar(full_phase, full_flux,
                     yerr=full_flux_err,
                     marker=filter_symbols[band], color=filter_colors[band], linestyle='none',
                     markersize=7, fillstyle='none')

    
        ax2.yaxis.tick_right()
        ax2.yaxis.set_label_position('right')
        
        ax1.minorticks_on()
        ax2.minorticks_on()
        ax1.legend(ncols=2,loc="upper left")
        ax1.set_ylabel('PSF difference \nimage flux (nJy)',fontsize=14)
        ax2.set_ylabel('PSF direct\nimage flux (nJy)',rotation=-90,labelpad=35,fontsize=14)
        ax2.set_xlabel('Phase',fontsize=14)
        ax1.set_xlabel('Phase',fontsize=14)

        ax1.set_title(f"{band}-band DiffFlux Phase Curve (Filtered)",fontsize=16)
        ax2.set_title(f"{band}-band Flux Phase Curve (Filtered)",fontsize=16)

        if len(masked_diff):
            ax1.set_ylim(np.min(masked_diff) - 2e4, np.max(masked_diff) + 2e4)
            ax2.set_ylim(np.min(masked_flux) - 2e4, np.max(masked_flux) + 2e4)
    
        ax1.set_xlim(-0.05,2.05)
        ax2.set_xlim(-0.05,2.05)
    
    plt.tight_layout()
    plt.suptitle(f"DiaObjectId - {diaId}",fontsize=20,y=1.01)
    
    if directory:
        plt.savefig(f"{directory}/combined_filt_flux.png",dpi=330,bbox_inches="tight")
        pass
    else:
        plt.show()
    plt.close()



"""
================================================================================================================
Overall Above ^

Individual Below v
================================================================================================================
"""


def individ_filt_phase_flux_plots(star_group,diaId,phase,directory="",filt=2.5):
    fig, ax = plt.subplots(1, 2, figsize=(14, 4))
    fig.subplots_adjust(wspace=0)

    max_diff = float("-inf")
    min_diff = float("inf")
    
    max_flux = float("-inf")
    min_flux = float("inf")
    
    ax1 = ax[0]
    ax2 = ax[1]
    
    for band in filter_names:
        pickband = (star_group['band'] == band) #& (star_group['psfDiffFlux'] > -50000)
    
        temp_diff = star_group[pickband]['psfDiffFlux']
        temp_flux = star_group[pickband]['psfFlux']
    
        if len(temp_flux):
            med = np.median(temp_diff)
            std3 = np.std(temp_diff) * filt
        else:
            med = 0
            std3 = 0
    
        mask = (temp_diff > med - std3) & (temp_diff < med + std3)
    
        masked_diff = temp_diff[mask]
        masked_flux = temp_flux[mask]
    
        full_phase = pd.concat([phase[pickband][mask],phase[pickband][mask] + 1])
        full_diff = pd.concat([masked_diff,masked_diff])
        full_diff_err = pd.concat([star_group[pickband]['psfDiffFluxErr'][mask],star_group[pickband]['psfDiffFluxErr'][mask]])
    
        full_flux = pd.concat([masked_flux,masked_flux])
        full_flux_err = pd.concat([star_group[pickband]['psfFluxErr'][mask],star_group[pickband]['psfFluxErr'][mask]])
        
        ax1.errorbar(full_phase, full_diff,
                 yerr=full_diff_err,
                 marker=filter_symbols[band], color=filter_colors[band],
                 label=f"{band}", markersize=7, linestyle='none', fillstyle='none')
        ax2.errorbar(full_phase, full_flux,
                     yerr=full_flux_err,
                     marker=filter_symbols[band], color=filter_colors[band], linestyle='none',
                     markersize=7, fillstyle='none')
    
    
        if np.max(masked_diff) > max_diff:
            max_diff = np.max(masked_diff)
    
        if np.min(masked_diff) < min_diff:
            min_diff = np.min(masked_diff) 
    
        if np.max(masked_flux) > max_flux:
            max_flux = np.max(masked_flux)
    
        if np.min(masked_flux) < min_flux:
            min_flux = np.min(masked_flux) 
    
    ax2.yaxis.tick_right()
    ax2.yaxis.set_label_position('right')
    
    ax1.minorticks_on()
    ax2.minorticks_on()
    ax1.legend(ncols=2,loc="upper left")
    ax1.set_ylabel('PSF difference\nimage flux (nJy)',fontsize=14)
    ax2.set_ylabel('PSF direct\nimage flux (nJy)',rotation=-90,labelpad=35,fontsize=14)
    ax2.set_xlabel('Phase',fontsize=14)
    ax1.set_xlabel('Phase',fontsize=14)

    ax1.set_title(f"Overall DiffFlux Phase Curve (Filtered)",fontsize=16)
    ax2.set_title(f"Overall Flux Phase Curve (Filtered)",fontsize=16)

    try:
        ax1.set_ylim(min_diff - 2e4, max_diff + 2e4)
        ax2.set_ylim(min_flux - 2e4, max_flux + 2e4)
    except:
        pass
    
    ax1.set_xlim(-0.05,2.05)
    ax2.set_xlim(-0.05,2.05)

    plt.suptitle(f"DiaObjectId - {diaId}",fontsize=20)
    plt.tight_layout()
    
    if directory:
        plt.savefig(f"{directory}/overall_filt_flux.png",dpi=330,bbox_inches="tight")
        pass
    else:
        plt.show()
    plt.close()
    
    for i in range(1,7):
        fig, ax = plt.subplots(1, 2, figsize=(14, 4))
        fig.subplots_adjust(wspace=0)
        
        ax1 = ax[0]
        ax2 = ax[1]
    
        band = filter_names[i-1]
    
        pickband = (star_group['band'] == band) #& (star_group['psfDiffFlux'] > -50000)
    
        temp_diff = star_group[pickband]['psfDiffFlux']
        temp_flux = star_group[pickband]['psfFlux']
    
        if len(temp_flux):
            med = np.median(temp_diff)
            std3 = np.std(temp_diff) * filt
        else:
            med = 0
            std3 = 0
    
        mask = (temp_diff > med - std3) & (temp_diff < med + std3)
    
        masked_diff = temp_diff[mask]
        masked_flux = temp_flux[mask]
    
        full_phase = pd.concat([phase[pickband][mask],phase[pickband][mask] + 1])
        full_diff = pd.concat([masked_diff,masked_diff])
        full_diff_err = pd.concat([star_group[pickband]['psfDiffFluxErr'][mask],star_group[pickband]['psfDiffFluxErr'][mask]])
    
        full_flux = pd.concat([masked_flux,masked_flux])
        full_flux_err = pd.concat([star_group[pickband]['psfFluxErr'][mask],star_group[pickband]['psfFluxErr'][mask]])
        
        ax1.errorbar(full_phase, full_diff,
                 yerr=full_diff_err,
                 marker=filter_symbols[band], color=filter_colors[band],
                 label=f"{band}", markersize=7, linestyle='none', fillstyle='none')
        ax2.errorbar(full_phase, full_flux,
                     yerr=full_flux_err,
                     marker=filter_symbols[band], color=filter_colors[band], linestyle='none',
                     markersize=7, fillstyle='none')

    
        ax2.yaxis.tick_right()
        ax2.yaxis.set_label_position('right')
        
        ax1.minorticks_on()
        ax2.minorticks_on()
        ax1.legend(ncols=2,loc="upper left")
        ax1.set_ylabel('PSF difference \nimage flux (nJy)',fontsize=14)
        ax2.set_ylabel('PSF direct\nimage flux (nJy)',rotation=-90,labelpad=35,fontsize=14)
        ax2.set_xlabel('Phase',fontsize=14)
        ax1.set_xlabel('Phase',fontsize=14)

        ax1.set_title(f"{band}-band DiffFlux Phase Curve (Filtered)",fontsize=16)
        ax2.set_title(f"{band}-band Flux Phase Curve (Filtered)",fontsize=16)

        if len(masked_diff):
            ax1.set_ylim(np.min(masked_diff) - 2e4, np.max(masked_diff) + 2e4)
            ax2.set_ylim(np.min(masked_flux) - 2e4, np.max(masked_flux) + 2e4)
    
        ax1.set_xlim(-0.05,2.05)
        ax2.set_xlim(-0.05,2.05)

        plt.suptitle(f"DiaObjectId - {diaId}",fontsize=20)
        plt.tight_layout()
        
        if directory:
            plt.savefig(f"{directory}/{band}_filt_flux.png",dpi=330,bbox_inches="tight")
            pass
        else:
            plt.show()
        plt.close()
```

```python
def phase_mag_plots(star_group,diaId,phase,mag,directory=""):
    
    fig, ax = plt.subplots(7, 1, figsize=(7, 21))
    
    ax1 = ax[0]
    for band in filter_names:
        pickband = star_group['band'] == band
    
        FluxErr = star_group[pickband]["psfFluxErr"]
        FluxBand = star_group[pickband]["psfFlux"]
    
        #error prop taken from 
        # https://astronomy.stackexchange.com/questions/38371/how-can-i-calculate-the-uncertainties-in-magnitude-like-the-cds-does
        yerr = (2.5 / np.log(10)) * (FluxErr / FluxBand)
    
        full_phase = pd.concat([phase[pickband],phase[pickband] + 1])
        full_mag = pd.concat([mag[pickband],mag[pickband]])
        full_mag_err = pd.concat([yerr,yerr])
        
        ax1.errorbar(full_phase, full_mag, yerr=full_mag_err,
                marker=filter_symbols[band], color=filter_colors[band],
                label=f"{band}", markersize=7, linestyle='none', fillstyle='none')
    
    ax1.set_ylabel('PSF magnitude',fontsize=14)
    ax1.set_xlabel('Phase',fontsize=14)
    ax1.set_title(f"Overall Mag Phase Curve (Unfiltered)",fontsize=16)
    ax1.minorticks_on()
    ax1.legend(ncols=2, loc='lower left')
    ax1.invert_yaxis()
    ax1.set_xlim(-0.05,2.05)
    
    for i in range(1,7):
        ax1 = ax[i]
    
        band = filter_names[i-1]
    
        pickband = star_group['band'] == band
        temp_mag = mag[pickband]
    
        FluxErr = star_group[pickband]["psfFluxErr"]
        FluxBand = star_group[pickband]["psfFlux"]
    
        yerr = (2.5 / np.log(10)) * (FluxErr / FluxBand)
    
        full_phase = pd.concat([phase[pickband],phase[pickband] + 1])
        full_mag = pd.concat([mag[pickband],mag[pickband]])
        full_mag_err = pd.concat([yerr,yerr])
    
        
        ax1.errorbar(full_phase, full_mag, yerr=full_mag_err,
                marker=filter_symbols[band], color=filter_colors[band],
                label=f"{band}", markersize=7, linestyle='none', fillstyle='none')
        
        ax1.set_ylabel('PSF magnitude',fontsize=14)
        ax1.set_xlabel('Phase',fontsize=14)
        ax1.set_title(f"{band}-band Mag Phase Curve (Unfiltered)",fontsize=16)
        ax1.minorticks_on()
        ax1.legend(ncols=2, loc='lower left')
        ax1.invert_yaxis()

        ax1.set_xlim(-0.05,2.05)
        
    plt.tight_layout()
    plt.suptitle(f"DiaObjectId - {diaId}",fontsize=20,y=1.01)

    if directory:
        plt.savefig(f"{directory}/combined_mag.png",dpi=330,bbox_inches="tight")
        pass
    else:
        plt.show()
    plt.close()




"""
================================================================================================================
Overall Above ^

Individual Below v
================================================================================================================
"""

def individ_phase_mag_plots(star_group,diaId,phase,mag,directory=""):

    fig, ax = plt.subplots(1, 1, figsize=(7, 4))
    
    ax1 = ax
    for band in filter_names:
        pickband = star_group['band'] == band
    
        FluxErr = star_group[pickband]["psfFluxErr"]
        FluxBand = star_group[pickband]["psfFlux"]
    
        #error prop taken from 
        # https://astronomy.stackexchange.com/questions/38371/how-can-i-calculate-the-uncertainties-in-magnitude-like-the-cds-does
        yerr = (2.5 / np.log(10)) * (FluxErr / FluxBand)
    
        full_phase = pd.concat([phase[pickband],phase[pickband] + 1])
        full_mag = pd.concat([mag[pickband],mag[pickband]])
        full_mag_err = pd.concat([yerr,yerr])
        
        ax1.errorbar(full_phase, full_mag, yerr=full_mag_err,
                marker=filter_symbols[band], color=filter_colors[band],
                label=f"{band}", markersize=7, linestyle='none', fillstyle='none')
    
    ax1.set_ylabel('PSF magnitude',fontsize=14)
    ax1.set_xlabel('Phase',fontsize=14)
    ax1.set_title(f"Overall Mag Phase Curve (Unfiltered)",fontsize=16)
    ax1.minorticks_on()
    ax1.legend(ncols=2, loc='lower left')
    ax1.invert_yaxis()
    ax1.set_xlim(-0.05,2.05)

    plt.suptitle(f"DiaObjectId - {diaId}",fontsize=20)
    plt.tight_layout()

    if directory:
        plt.savefig(f"{directory}/overall_mag.png",dpi=330,bbox_inches="tight")
        pass
    else:
        plt.show()
    plt.close()
    
    for i in range(1,7):
        fig, ax = plt.subplots(1, 1, figsize=(7, 4))
        ax1 = ax
    
        band = filter_names[i-1]
    
        pickband = star_group['band'] == band
        temp_mag = mag[pickband]
    
        FluxErr = star_group[pickband]["psfFluxErr"]
        FluxBand = star_group[pickband]["psfFlux"]
    
        yerr = (2.5 / np.log(10)) * (FluxErr / FluxBand)
    
        full_phase = pd.concat([phase[pickband],phase[pickband] + 1])
        full_mag = pd.concat([mag[pickband],mag[pickband]])
        full_mag_err = pd.concat([yerr,yerr])
    
        
        ax1.errorbar(full_phase, full_mag, yerr=full_mag_err,
                marker=filter_symbols[band], color=filter_colors[band],
                label=f"{band}", markersize=7, linestyle='none', fillstyle='none')
        
        ax1.set_ylabel('PSF magnitude',fontsize=14)
        ax1.set_xlabel('Phase',fontsize=14)
        ax1.set_title(f"{band}-band Mag Phase Curve (Unfiltered)",fontsize=16)
        ax1.minorticks_on()
        ax1.legend(ncols=2, loc='lower left')
        ax1.invert_yaxis()

        ax1.set_xlim(-0.05,2.05)

        plt.suptitle(f"DiaObjectId - {diaId}",fontsize=20)
        plt.tight_layout()
    
        if directory:
            plt.savefig(f"{directory}/{band}_mag.png",dpi=330,bbox_inches="tight")
            pass
        else:
            plt.show()
        plt.close()
```

## Defining Set-Up Functions

```python
def find_period(star_row):
    classif = star_row["DR3_best_classification"]

    if classif == "RRab":
        period_p = star_row["DR3_pf"]
        period_s = "N/A"
    elif classif == "RRc":
        period_p = star_row["DR3_p1_o"]
        period_s = "N/A"
    elif classif == "RRd":
        period_p = star_row["DR3_p1_o"]
        period_s = star_row["DR3_pf"]
    else:
        print("Star has no best classification.")
        period_p = star_row["DR3_pf"]
        period_s = star_row["DR3_p1_o"]

    return period_p,period_s
```

```python
def find_phase(group_data,period_p,filt=2.5):
    band_dict = {}
    for band in filter_names:
        band_dict[band] = np.sum((group_data['band'] == band))

    max_band = sorted(band_dict.items(),key=lambda x:x[-1])[-1][0]

    pickr = group_data["band"] == max_band

    temp_flux = group_data["psfFlux"][pickr]

    #need to update if change filtering method
    med = np.median(temp_flux)
    std3 = np.std(temp_flux) * filt

    mask = (temp_flux > med - std3) & (temp_flux < med + std3)

    filt_flux = temp_flux[mask]

    ind_max = filt_flux.idxmax() #find max to use as start point!

    t0 = group_data['expMidptMJD'][pickr].loc[ind_max]

    mjd_norm = (group_data['expMidptMJD'] - t0) / period_p

    phase = np.mod(mjd_norm, 1.0)

    return t0, phase
```

```python
def convert_to_mag(group_data):
    mag = -2.5 * np.log10(group_data['psfFlux'] / (3631 * 1e9))
    return mag
```

```python
def create_plots(row,path="",filt=2.5):
    print(f"Making lightcurve plots for {row["diaObjectId"]}...")

    #pull out forcedDiaSources associated with current star
    group_data = grouped_forcedDia.get_group(row["diaObjectId"])
    #filted out any negative flux values
    group_data = group_data[group_data['psfFlux'] > 0]

    #mask to remove all bad pixels or saturated measurements from sources
    back_sat = ~group_data["pixelFlags_saturated"]
    cent_sat = ~group_data["pixelFlags_saturatedCenter"]
    bad_pixel = ~group_data["pixelFlags_bad"]

    #update group data based on mask
    group_data = group_data[back_sat & cent_sat & bad_pixel]

    #find period for star
    period_p,period_s = find_period(row)

    #save star diaObjectId
    diaId = row["diaObjectId"]

    if path:
        # Creating New Folder for current star
        dia_path = os.path.join(path,f"{diaId}")
        
        if not os.path.isdir(dia_path):
            os.mkdir(dia_path)

    """
    Plot/Save Flux Curves over MJD
    """

    if path:
        # Creating New Folder for days plots
        days_path = os.path.join(dia_path,"days")
        
        if not os.path.isdir(days_path):
            os.mkdir(days_path)
            
    days_flux_plots(group_data,diaId,directory=f"{days_path}")
    
    individ_days_flux_plots(group_data,diaId,directory=f"{days_path}")

    #finding 0 time and phase from data
    t0, phase = find_phase(group_data,period_p,filt=filt)

    """
    Plot/Save Flux Curves over 2 Phases (unfiltered)
    """

    if path:
        # Creating New Folder for days plots
        flux_path = os.path.join(dia_path,"phase")
        
        if not os.path.isdir(flux_path):
            os.mkdir(flux_path)
    
    phase_flux_plots(group_data,diaId,phase,directory=f"{flux_path}")

    individ_phase_flux_plots(group_data,diaId,phase,directory=f"{flux_path}")

    """
    Plot/Save Flux Curves over 2 Phases (filtered)
    """

    if path:
        # Creating New Folder for days plots
        filt_flux_path = os.path.join(dia_path,"filt_phase")
        
        if not os.path.isdir(filt_flux_path):
            os.mkdir(filt_flux_path)

    filt_phase_flux_plots(group_data,diaId,phase,directory=f"{filt_flux_path}",filt=filt)

    individ_filt_phase_flux_plots(group_data,diaId,phase,directory=f"{filt_flux_path}",filt=filt)

    #find mag for group data
    mag = convert_to_mag(group_data)

    """
    Plot/Save Mag Curves over 2 Phases (unfiltered)
    """

    if path:
        # Creating New Folder for days plots
        mag_path = os.path.join(dia_path,"mag")
        
        if not os.path.isdir(mag_path):
            os.mkdir(mag_path)
    
    phase_mag_plots(group_data,diaId,phase,mag,directory=f"{mag_path}")

    individ_phase_mag_plots(group_data,diaId,phase,mag,directory=f"{mag_path}")

    """
    Plot/Save Mag Curves over 2 Phases (filtered)
    """

    if path:
        # Creating New Folder for days plots
        filt_mag_path = os.path.join(dia_path,"filt_mag")
        
        if not os.path.isdir(filt_mag_path):
            os.mkdir(filt_mag_path)
    
    filt_phase_mag_plots(group_data,diaId,phase,mag,directory=f"{filt_mag_path}",filt=filt)

    individ_filt_phase_mag_plots(group_data,diaId,phase,mag, directory=f"{filt_mag_path}",filt=filt)


    pass
```
