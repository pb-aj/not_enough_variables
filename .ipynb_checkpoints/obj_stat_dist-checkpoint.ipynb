# Object Cross Match Distributions

The following notebook creates various plots to show matched stars distributions through the Object cross search.  The results of these plots can be found in obj_stat_plots and the draft (for diaObj) can be found in test_stat_dist.ipynb.

## Loading in Python Libraries and Needed Data 

```python
#import libraries and set up tap service
import matplotlib.pyplot as plt
import numpy as np
import astropy.units as u
from lsst.rsp import get_tap_service
import pandas as pd
import os
import sys
pd.set_option('display.max_columns', None)

from lsst.utils.plotting import (get_multiband_plot_colors,
                                 get_multiband_plot_symbols)

from lsst.daf.butler import Butler
from lsst.utils.plotting import (
    get_multiband_plot_colors,
    get_multiband_plot_symbols,
    get_multiband_plot_linestyles
)

#setting up butler
butler = Butler('dp1', collections='LSSTComCam/DP1')
assert butler is not None

import seaborn as sns
my_pal = sns.color_palette("colorblind")
```

```python
# Creating New Folder for obj_stat_plots
outdir = os.getcwd()
stat_path = os.path.join(outdir,"obj_stat_plots")

if not os.path.isdir(stat_path):
    os.mkdir(stat_path)
```

```python
Obj_df = pd.read_csv("data/Obj_df.csv",index_col=0)
forcedObj_df = pd.read_csv("data/ObjForced_df.csv",index_col=0)

grouped_forced = forcedObj_df.groupby("ObjectId")
```

```python
Obj_df.shape
```

```python
hue_ord = list(Obj_df["DR3_field_loc_center"].unique())
hue_ord.sort()
```

```python
#function to pull out period for each star regardless of classification
#for RRd's, use p1_o as this is more common for fitting
def get_p(row):

    if row["DR3_best_classification"] == "RRab":
        return row["DR3_pf"]
    else:
        return row["DR3_p1_o"]


ps = Obj_df.apply(get_p,axis=1)
```

```python
filter_names = ['u', 'g', 'r', 'i', 'z', 'y']
filter_colors = get_multiband_plot_colors()
filter_symbols = get_multiband_plot_symbols()
filter_linestyles = get_multiband_plot_linestyles()
```

```python
#adding in center point of fields from 301_0_DP1
field_centers = {
    "47 Tuc": [6.02, -72.08],
    "SV 38 7": [37.86, 6.98],
    "Fornax": [40.00, -34.45],
    "ECDFS": [53.13, -28.10],
    "EDFS": [59.10, -48.73],
    "SV 95 -25": [95.00, -25.00],
    "Seagull": [106.23, -10.51],
}
field_names = list(field_centers.keys())
```

```python
#load in the maglim for each object
u_maglim = butler.get('deepCoadd_psf_maglim_consolidated_map_weighted_mean',
                      band='u')
g_maglim = butler.get('deepCoadd_psf_maglim_consolidated_map_weighted_mean',
                      band='g')
r_maglim = butler.get('deepCoadd_psf_maglim_consolidated_map_weighted_mean',
                      band='r')
i_maglim = butler.get('deepCoadd_psf_maglim_consolidated_map_weighted_mean',
                      band='i')
z_maglim = butler.get('deepCoadd_psf_maglim_consolidated_map_weighted_mean',
                      band='z')
y_maglim = butler.get('deepCoadd_psf_maglim_consolidated_map_weighted_mean',
                      band='y')
```

## Saving each figure above as an individual plot

### Coord Plots

```python
#loop through each field and save a plot of mag lim with matched stars overlayed
for i in range(len(hue_ord)):
    name = hue_ord[i]
    ra_cen, dec_cen = field_centers[name]
    span_dec = 0.75
    if name == 'SV 38 7':
        span_dec = 1.00
    span_ra = span_dec / np.cos(np.deg2rad(dec_cen))
    ra = np.linspace(ra_cen-span_ra, ra_cen+span_ra, 250)
    dec = np.linspace(dec_cen-span_dec, dec_cen+span_dec, 250)

    x, y = np.meshgrid(ra, dec)
    values = r_maglim.get_values_pos(x, y)
    row, col = np.where(values < 0.0)
    values[row, col] = 'NaN'
    plt.pcolormesh(x, y, values, vmin=24.0, vmax=27)
    plt.colorbar(label="PSF Mag Limit (r-band)")

    size = 20

    conversion_mas_deg = 1 / 3600000

    current_stars = Obj_df[Obj_df["DR3_field_loc_center"] == name]
    plt.errorbar(current_stars["DR3_ra"],current_stars["DR3_dec"],current_stars["DR3_ra_error"] * conversion_mas_deg,
                 current_stars["DR3_dec_error"] * conversion_mas_deg,fmt="none",elinewidth=size/2,c="k")
    plt.scatter(current_stars["DR3_ra"],current_stars["DR3_dec"],s=size,c="k")

    plt.xlabel("Right Ascension (deg)",fontsize=16)
    plt.ylabel("Declination (deg)",fontsize=16)
    plt.title(name,fontsize=20)
    plt.gca().invert_xaxis()
    plt.tight_layout()
    plt.savefig(f"{stat_path}/{name}_field_RRLs.png",dpi=330,bbox_inches="tight")
    plt.show()
    plt.close()
```

### Saving per-band Distribution Plots

```python
#save each n[x] band distribution by looping through each

for i in range(len(["nu","ng","nr","ni","nz","ny"])):
    filt = ["nu","ng","nr","ni","nz","ny"][i]
    p = sns.histplot(Obj_df, x=Obj_df[filt], hue="DR3_field_loc_center", 
                     fill=True, hue_order=hue_ord, multiple="stack",palette=my_pal[:len(hue_ord)],
                     legend=True,log_scale=True)

    bins_edges = [patch.get_x() for patch in p.patches] + [p.patches[-1].get_x() + p.patches[-1].get_width()]
    bins_edges = set(bins_edges)
    plt.xticks([e for e in bins_edges],labels=[f"{np.ceil(e):.0f}" for e in bins_edges])
    plt.tick_params(axis='x', which='minor', bottom=False)

    plt.xlabel(f"# of Observations",fontsize=16)
    plt.ylabel("Count",fontsize=16)
    plt.title(f"{filt[-1]}-band Distribution",fontsize=20)
    sns.move_legend(p, "lower center", bbox_to_anchor=(1.2, .5),title="")

    plt.tight_layout()
    plt.savefig(f"{stat_path}/{filt[-1]}_band_dist.png",dpi=330,bbox_inches="tight")
    plt.show()
    plt.close()
```

## Saving Additional Plots

### Mag Mag Comparison

```python
#find the r and g average magnitude after converting flux through AB mag
def find_r_group_avg(row):
    ObjID = row["ObjectId"]
    group = grouped_forced.get_group(ObjID)

    flux_vals = group[group["band"] == "r"]["psfFlux"]
    flux_vals = flux_vals[flux_vals > 0]
    mag = -2.5 * np.log10( flux_vals / (3631 * 1e9))

    return np.mean(mag)


def find_g_group_avg(row):
    ObjID = row["ObjectId"]
    group = grouped_forced.get_group(ObjID)

    flux_vals = group[group["band"] == "g"]["psfFlux"]
    flux_vals = flux_vals[flux_vals > 0]
    mag = -2.5 * np.log10( flux_vals / (3631 * 1e9))

    return np.mean(mag)
```

```python
#call above functions and save new columns to use
mean_mags_r = Obj_df.apply(find_r_group_avg,axis=1)
mean_mags_g = Obj_df.apply(find_g_group_avg,axis=1)

Obj_df["mean_r_mag"] = mean_mags_r
Obj_df["mean_g_mag"] = mean_mags_g
```

```python
#plot mag in Rubin vs GAIA to check for outliers
fig, ax = plt.subplots()

p = sns.scatterplot(Obj_df, x = Obj_df["DR3_phot_g_mean_mag"], y = Obj_df["mean_r_mag"],
            style = Obj_df["DR3_field_loc_center"], style_order = hue_ord,
            legend=True,hue=Obj_df["nr"],palette = "vlag")

sns.lineplot(x = np.linspace(np.min(Obj_df["mean_r_mag"]), np.max(Obj_df["mean_r_mag"]),1000),
         y = np.linspace(np.min(Obj_df["mean_r_mag"]), np.max(Obj_df["mean_r_mag"]),1000),
         c = "k",alpha=.5,linestyle=":",label="1:1 Line",legend=True)

handles, labels = p.get_legend_handles_labels()

norm = plt.Normalize(Obj_df["nr"].min(), Obj_df["nr"].max())
sm = plt.cm.ScalarMappable(cmap="vlag", norm=norm)
sm.set_array([])

plt.xlabel("Average G Mag from GAIA",fontsize=16)
plt.ylabel("Average R Mag from Rubin",fontsize=16)
plt.title("R Mag in Rubin vs G Mag in GAIA",fontsize=20)
plt.legend(handles=handles[-6:], labels=labels[-6:])

cax = fig.add_axes([0.92, 0.15, 0.03, 0.7])
p.figure.colorbar(sm,cax=cax)
plt.savefig(f"{stat_path}/mag_mag_comp.png",dpi=330,bbox_inches="tight")
plt.show()
plt.close()
```

### Low Epoch Checks

```python
#create a column that filters each row on if it has over 10 epochs of observation
Obj_df["over_10_epochs"] = Obj_df["nSources"] > 10
```

```python
#plot the same Coord plots color coded by over_10_epochs 
for i in range(len(hue_ord)):
    name = hue_ord[i]
    ra_cen, dec_cen = field_centers[name]
    span_dec = 0.75
    if name == 'SV 38 7':
        span_dec = 1.00
    span_ra = span_dec / np.cos(np.deg2rad(dec_cen))
    ra = np.linspace(ra_cen-span_ra, ra_cen+span_ra, 250)
    dec = np.linspace(dec_cen-span_dec, dec_cen+span_dec, 250)

    x, y = np.meshgrid(ra, dec)
    values = r_maglim.get_values_pos(x, y)
    row, col = np.where(values < 0.0)
    values[row, col] = 'NaN'
    plt.pcolormesh(x, y, values, vmin=24.0, vmax=27)
    plt.colorbar(label="PSF Mag Limit (r-band)")
    
    size = 20

    conversion_mas_deg = 1 / 3600000

    current_stars = Obj_df[Obj_df["DR3_field_loc_center"] == name]
    c_map = current_stars["over_10_epochs"].map({True:"aqua",False:"magenta"})
    mask = Obj_df["over_10_epochs"]

    plt.scatter(current_stars["DR3_ra"][mask],current_stars["DR3_dec"][mask],s=size,
                color="aqua", edgecolors='k',label=">10 Epochs")
    plt.scatter(current_stars["DR3_ra"][~mask],current_stars["DR3_dec"][~mask],s=size,
                color="magenta", edgecolors='k',label="<=10 Epochs")

    
    plt.errorbar(current_stars["DR3_ra"],current_stars["DR3_dec"],current_stars["DR3_ra_error"] * conversion_mas_deg,
                 current_stars["DR3_dec_error"] * conversion_mas_deg,fmt="none",elinewidth=size/2,
                 ecolor=c_map)

    plt.xlabel("Right Ascension (deg)",fontsize=16)
    plt.ylabel("Declination (deg)",fontsize=16)
    plt.title(name,fontsize=20)
    plt.gca().invert_xaxis()
    plt.legend()
    plt.tight_layout()
    plt.savefig(f"{stat_path}/{name}_colored_field_RRLs.png",dpi=330,bbox_inches="tight")
    plt.show()
    plt.close()
```

```python
#find distance of each star to its field center
def find_distance_to_center_field(row):
    current_field = row["DR3_field_loc_center"]
    center_coords = field_centers[current_field]

    star_coords = [row["ra"],row["dec"]]

    distance = np.sqrt((star_coords[0] - center_coords[0])**2 + \
                       (star_coords[1] - center_coords[1])**2)
    return distance

Obj_df["dist_to_field_center"] = Obj_df.apply(find_distance_to_center_field,axis=1)
```

```python
#plot the distribution found above against # of epochs
fig, ax = plt.subplots()

p = sns.scatterplot(Obj_df, y = Obj_df["dist_to_field_center"], x = Obj_df["nSources"],
            hue = Obj_df["DR3_field_loc_center"], hue_order = hue_ord,
            legend=True)

handles, labels = p.get_legend_handles_labels()

plt.ylabel("Distance From Field's Center",fontsize=16)
plt.xlabel("Number of Epochs Observed",fontsize=16)
plt.xscale("log")
plt.title("Distance from Center versus # of Epochs",fontsize=20)
plt.legend(handles=handles[1:], labels=labels[1:])
plt.savefig(f"{stat_path}/dist_v_epoch_scatter.png",dpi=330,bbox_inches="tight")
plt.show()
plt.close()
```

```python
#plot it against mean mag in r
fig, ax = plt.subplots()

p = sns.scatterplot(Obj_df, y = Obj_df["mean_r_mag"], x = Obj_df["nSources"],
            style = Obj_df["DR3_field_loc_center"], style_order = hue_ord,
            legend=True,hue=Obj_df["nr"],palette = "vlag")

handles, labels = p.get_legend_handles_labels()

plt.ylabel("Mean R Mag (From Rubin)",fontsize=16)
plt.xlabel("Number of Epochs Observed",fontsize=16)
plt.xscale("log")
plt.title("Mean R Mag versus # of Epochs",fontsize=20)
plt.legend(handles=handles[-5:], labels=labels[-5:])

cax = fig.add_axes([0.92, 0.15, 0.03, 0.7])
p.figure.colorbar(sm,cax=cax)
plt.savefig(f"{stat_path}/Mean_r_v_epoch_scatter.png",dpi=330,bbox_inches="tight")
plt.show()
plt.close()
```

### Duration Distribution Plots

```python
def find_epoch_duration_per_band(row):

    ObjID = row["ObjectId"]
    Group = grouped_forced.get_group(ObjID)

    diff_l = []

    for band in filter_names:
        rows_band = Group[Group["band"] == band]
        first_date = np.min(rows_band["expMidptMJD"])
        last_date = np.max(rows_band["expMidptMJD"])

        diff = last_date - first_date

        diff_l.append(diff)

    return pd.Series(diff_l)
```

```python
Obj_df[[f"{b}_duration" for b in filter_names]] = Obj_df.apply(find_epoch_duration_per_band,axis=1)
```

```python
fig, ax = plt.subplots()

for band in filter_names:
    p = plt.scatter(Obj_df[f"n{band}"], Obj_df[f"{band}_duration"],
                        marker=filter_symbols[band], color=filter_colors[band],label=f"{band}")

# handles, labels = p.get_legend_handles_labels()

plt.xlabel("Number of Epochs (by band)",fontsize=16)
plt.ylabel("Time Between First & Last Epoch",fontsize=16)
plt.title("Epoch Duration vs. Number of Epochs (by band)",fontsize=20)
plt.legend()
plt.savefig(f"{stat_path}/duration_by_band.png",dpi=330,bbox_inches="tight")
plt.show()
plt.close()
```
